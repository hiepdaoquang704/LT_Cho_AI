```python

def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):

	n_vars = 1 if type(data) is list else data.shape[1]
	df = pd.DataFrame(data)
	cols, names = list(), list()

	for i in range(n_in, 0, -1):
		cols.append(df.shift(i))
		names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]

	for i in range(0, n_out):
		cols.append(df.shift(-i))
		if i == 0:
			names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
		else:
			names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]

	agg = pd.concat(cols, axis=1)
	agg.columns = names

	if dropnan:
		agg.dropna(inplace=True)
	return agg
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.multioutput import MultiOutputRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import GridSearchCV

# Định nghĩa các mô hình và tham số để tìm kiếm
models = {
    'LinearRegression': MultiOutputRegressor(LinearRegression()),
    'RandomForestRegressor': MultiOutputRegressor(RandomForestRegressor()),
    'SVR': MultiOutputRegressor(Pipeline([('scaler', StandardScaler()), ('svr', SVR())]))
}

params = {
    'LinearRegression': {},
    'RandomForestRegressor': {
        'estimator__n_estimators': [50, 100, 200],
        'estimator__max_depth': [None, 10, 20]
    },
    'SVR': {
        'estimator__svr__C': [0.1, 1, 10],
        'estimator__svr__gamma': ['scale', 'auto']
    }
}

# Tìm kiếm tham số tốt nhất cho từng mô hình
best_models = {}
for model_name in models:
    grid = GridSearchCV(models[model_name], params[model_name], scoring='neg_mean_squared_error', cv=3)
    grid.fit(X_train, y_train)
    best_models[model_name] = grid.best_estimator_

# Kết quả tìm kiếm
for model_name in best_models:
    print(f"Best parameters for {model_name}: {best_models[model_name]}")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Hàm tính MAPE
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Đánh giá các mô hình
results = {}
for model_name, model in best_models.items():
    y_pred = model.predict(X_test)
    results[model_name] = {
        'RMSE': np.sqrt(mean_squared_error(y_test, y_pred)),
        'MAE': mean_absolute_error(y_test, y_pred),
        'R2': r2_score(y_test, y_pred),
        'MAPE': mean_absolute_percentage_error(y_test, y_pred)
    }

# Lập bảng kết quả so sánh
results_df = pd.DataFrame(results).T
print(results_df)
```
