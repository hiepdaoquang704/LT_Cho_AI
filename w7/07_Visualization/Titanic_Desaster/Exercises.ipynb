### Step 1. Import the necessary libraries

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv)

### Step 3. Assign it to a variable titanic 

```python
url='https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv'
```

```python
titanic = pd.read_csv(url)

titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId').head()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
gender_counts = titanic['Sex'].value_counts()

# Plotting
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Male/Female Proportion')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
colors = {'male': 'skyblue', 'female': 'lightcoral'}


plt.figure(figsize=(10, 6))


plt.scatter(titanic[titanic['Sex'] == 'male']['Age'], titanic[titanic['Sex'] == 'male']['Fare'], color=colors['male'], label='Male', alpha=0.7)


plt.scatter(titanic[titanic['Sex'] == 'female']['Age'], titanic[titanic['Sex'] == 'female']['Fare'], color=colors['female'], label='Female', alpha=0.7)


plt.title('Scatter Plot of Fare vs Age (Colored by Gender)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

### Step 8. Create a histogram with the Fare payed

```python
plt.figure(figsize=(10, 6))
plt.hist(titanic['Fare'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Fare Paid')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
